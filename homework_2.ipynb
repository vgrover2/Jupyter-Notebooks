Use the code below to obtain the Flights data set. It contains information about distance traveled by the flights and delay in their arrival.

You might observe that some flights have a negative delay. In this question, our objective is to analyze the relationship between the `distance` and `delay` variables for all flights that arrived early (i.e. **delay < 0**).

Here is a hint:
1. Obtain a dataframe that contains all records with a negative delay (< 0).
2. Plot the `distance` and `delay` variables which represent the distance traveled by the flight and the delay in arrival to the destination.

Which of the statements would hold for the plot obtained?

A. There is a rough tendency that flights that travel a longer distance tend to arrive **later** (less negative delay).

B. There is a rough tendency that flights that travel a longer distance tend to arrive **earlier** (more negative delay).


<br>

**`Gold-Star Question`**: Think about what kind of plot is apt for this data. Does switching the x-axis and y-axis variable affect your conclusion?

````python
import pandas as pd
df = pd.read_csv("flights_20k.csv")
df
#df1.head(3)
````

````python
# your code here
import pandas as pd
df = pd.read_csv("flights_20k.csv")
df1 = df[df['delay'] > 0]
df2 = df[df['delay'] < 0]
#df2.head()
import matplotlib.pyplot as plt
%matplotlib inline

# Positive delay plot
distance = df1['distance']
delay = df1['delay']

plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.scatter(distance, delay, alpha=0.5)  # alpha controls point transparency

plt.xlabel('Distance (miles)')
plt.ylabel('Delay (minutes)')
plt.title('Scatter Plot of Distance vs. Delay')

plt.grid(True)
plt.show()
import matplotlib.pyplot as plt2
%matplotlib inline

# Negative delay plot
distance2 = df2['distance']
delay2 = df2['delay']

plt2.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt2.scatter(distance2, delay2, alpha=0.5)  # alpha controls point transparency

plt2.xlabel('Distance (miles)')
plt2.ylabel('Delay (minutes)')
plt2.title('Scatter Plot of Distance vs. Delay')

plt2.grid(True)
plt2.show()
````

````python
# your code here
import matplotlib.pyplot as plt
%matplotlib inline

budget = df3['Production_Budget']

plt.figure(figsize=(8, 5))  # Adjust the figure size as needed
plt.hist(budget, bins=15, color='skyblue', edgecolor='black')  # Adjust bins and colors

# Set labels and title
plt.xlabel('Production Budget')
plt.ylabel('Frequency')
plt.title('Histogram of Production Budget')

# Display the plot
plt.grid(True)
plt.show()

import matplotlib.pyplot as plt2 
%matplotlib inline

imdb = df3['IMDB_Rating']
usGross = df3['US_Gross']

plt2.figure(figsize = (8, 5))
plt2.scatter(imdb, usGross, alpha = 0.5)

plt2.xlabel('IMDB Rating')
plt2.ylabel('US_Gross')
plt2.title('Scatter of IMDB vs Gross')

plt2.grid(True)
plt2.show

import matplotlib.pyplot as plt3
%matplotlib inline
rottenTomatoes = df3['Rotten_Tomatoes_Rating']

plt3.figure(figsize = (8,5))
plt3.scatter(imdb, rottenTomatoes, alpha = 0.5)

plt3.xlabel('IMDB Rating')
plt3.ylabel('Rotten Tomatoes Rating')
plt3.title('IMDB vs Rotten Tomatoes')

plt3.grid(True)
plt3.show
````

````python
# your code here
df4 = pd.read_csv("gapminder.csv")

import matplotlib.pyplot as plt4
%matplotlib inline

lifeExpectancy = df4['life_expect']
fertility = df4['fertility']

plt4.figure(figsize = (8,5))
plt4.scatter(lifeExpectancy, fertility, alpha = 0.5)

plt4.xlabel('Life Expectancy')
plt4.ylabel('Fertility')
plt4.title('Life Expectancy vs Fertility')

plt4.grid(True)
plt4.show()

import matplotlib.pyplot as plt5
%matplotlib inline
df5 = df4[df4['country'] == 'Rwanda']
df5

year = df5['year']
population = df5['pop']

plt5.figure(figsize = (8, 5))
plt5.scatter(year, population, alpha = 0.5)

plt5.xlabel('Year')
plt5.ylabel('Population')
plt5.title('Population of Rwanda over time')

plt5.grid(True)
plt5.show()

import matplotlib.pyplot as plt6
%matplotlib inline
df6 = df4[df4['country'] == 'Kenya']
df6

year = df6['year']
population = df6['pop']

plt6.figure(figsize = (8, 5))
plt6.scatter(year, population, alpha = 0.5)

plt6.xlabel('Year')
plt6.ylabel('Population')
plt6.title('Population of Kenya over time')

plt6.grid(True)
plt6.show()
````
